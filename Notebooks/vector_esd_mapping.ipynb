# Exploratory Spatial Mapping of Vector Data

```python
from pathlib import Path
import os

# Assign the project root path to a variable:
project_root = Path.cwd().parent

# Change the directory to the project root level:
os.chdir(project_root)
```

```python
# Assign the GPKG dir to a variable:
gpkg_directory = project_root / "Data" / "GeoPackages"

# Create a dictionary of GPKG file names and pathways: 
gpkg_paths = {
    gpkg_path.stem: gpkg_path
    for gpkg_path in sorted(gpkg_directory.rglob("*.gpkg"))
}

# Print the GPKG file name (the dictionary key) and its relative path (the value):
for key, value in gpkg_paths.items():
    print(f'{key}:', value.relative_to(project_root))
```

```python
# Assign the GPKG file pathway to its respective variable:
Redmond_gpkg = gpkg_paths["Redmond"]
KingCounty_gpkg = gpkg_paths["KingCounty"]
```

```python
import geopandas as gpd

county_limit = gpd.read_file(
    KingCounty_gpkg,
    layer=0,
)

city_limit = gpd.read_file(
    Redmond_gpkg,
    layer="CityLimit",
)
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(12, 12))

county_limit.plot(
    ax=ax,
    facecolor="none",
    edgecolor="gray",
    linewidth=1,
)

city_limit.plot(
    ax=ax,
    facecolor="none",
    edgecolor="darkred",
    linewidth=1,
)

ax.set_title("Eastside Traffic Counts")
ax.set_axis_off()

plt.show()
```
